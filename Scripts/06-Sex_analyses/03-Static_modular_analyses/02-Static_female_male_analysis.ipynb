# Step 1: Compute communities via Louvain algorithm

```python
# Load data
mat_f = sio.loadmat('../../../Data/06-Sex_analyses/01-Adjacency_matrices/Females_adj_matrices.mat')
female_adj = mat_f['Females_adj_matrices']
mat_m = sio.loadmat('../../../Data/06-Sex_analyses/01-Adjacency_matrices/Males_adj_matrices.mat')
male_adj = mat_m['Males_adj_matrices']
```

```python
n_sub = 15 # number of subjects per group
n_roi = 98
n_freq = 6

# Form graphs
assigned_systems_f = np.zeros((n_sub,n_roi,n_freq))
assigned_systems_m = np.zeros((n_sub,n_roi,n_freq))

modularity_f = np.zeros((n_sub,n_freq))
modularity_m = np.zeros((n_sub,n_freq))

for sub in range(n_sub):
    print(sub)
    for freq in range(n_freq):
        # Convert matrix to a networkx graph (f+m)
        Female_mat_plot = female_adj[:,:,freq,sub]
        Male_mat_plot = male_adj[:,:,freq,sub]
        
        D_f = nx.Graph(Female_mat_plot)
        D_m = nx.Graph(Male_mat_plot)

        # compute the best partition and store the assigned systems in new array
        partition_f = community_louvain.best_partition(D_f)
        partition_m = community_louvain.best_partition(D_m)
        
        # Compute modularity of selected partition
        modularity_f[sub,freq] = community_louvain.modularity(partition_f, D_f)
        modularity_m[sub,freq] = community_louvain.modularity(partition_m, D_m)
        
        
        # Assign system
        system_affiliation_f = list(partition_f.values())
        system_affiliation_m = list(partition_m.values())
        
        assigned_systems_f[sub,:,freq] = system_affiliation_f
        assigned_systems_m[sub,:,freq] = system_affiliation_m
        
        
#np.save('../../../Data/06-Sex_analyses/03-Static_modular_analyses/modularity_f.npy', modularity_f)
#np.save('../../../Data/06-Sex_analyses/03-Static_modular_analyses/modularity_m.npy', modularity_m)
        
```

# Step 2a: Permutation test for modularity difference between males and females 

We now perform a permutation test with 10000 permutations

```python
# Mean across subjects
f_mean = modularity_f.mean(axis=0)
m_mean = modularity_m.mean(axis=0)
mean_dif = abs(f_mean - m_mean)
```

```python
n_perm = 10000
null_dist = np.zeros((n_perm,n_freq))
p_value = np.zeros((n_freq))

for freq in range(n_freq):
    print(freq)
    combined_data = np.concatenate((modularity_f[:,freq], modularity_m[:,freq])) # combine the data
    combined_data_copy = combined_data.copy()
    random.seed(123) # set seed for reproducibility

    # Perform randomization 10000 times
    for perm in range(n_perm):
        random.shuffle(combined_data_copy)
        shuffled_female = combined_data_copy[0:15]
        shuffled_male = combined_data_copy[15:30]
        null_dist[perm,freq] = abs(shuffled_female.mean(axis=0) - shuffled_male.mean(axis=0))
    
    # Calc number of permutations that are greater than the mean calculated earlier and divide this by number 
    # of permutations to get the p-values
    p_value[freq] = (sum(i > mean_dif[freq] for i in null_dist[:,freq]))/n_perm
    
print(p_value)
# Do we need to correct for multiple comparisons so divide by 6 because we are testing 6 frequency bands?
# Yes, if we use a Bonferri correction we simply multiple all p-values by 6
# How do we do a FDR correction?

```

### Account for false discovery rate

```python
# Organise p-value
p_value_ordered = np.sort(p_value)
idx_p_value = p_value.argsort()
print('p_values')
print(p_value)
print('p_values_ordered')
print(p_value_ordered)


freq_band = ['Delta', 'Theta', 'Alpha', 'Beta', 'Low Gamma', 'High Gamma']
freq_band_reordered = [freq_band[i] for i in idx_p_value]

print('freq_bands_reordered')
print(freq_band_reordered)

# Calculate critical value
crit_val = np.zeros((6,))
order = (p_value_ordered.argsort())+1
m = 6
Q = 0.05

for j in range(len(order)):
    i = order[j]
    crit_val[j] = (i/m)*Q

print('crit_values')
print(crit_val)

# Create data frame
crit_p_data = np.stack((p_value_ordered, order, crit_val))
np.transpose(crit_p_data)


df = pd.DataFrame(np.transpose(crit_p_data),index=freq_band_reordered, columns=['p-value','Rank','Crit val (I/m)*Q'])
df

# With a false discovert rate of 10%, only the low gamma is significant
# At a 5% rate, none of the results are significant
```

# Step 3: Permutation test for community structure difference between males and females 

We now perform a permutation test with 10,000 permutations. We begin by forming N_i, N_j, and N_ij.

### Function

```python
def NMI_func(A, B):
    # Note number of comunities in female and male partitions
    C_F = len(np.unique(A))
    C_M = len(np.unique(B))

    # Form N_i and N_j and store as array
    N_i = np.fromiter(Counter(A).values(), dtype=float)
    N_j = np.fromiter(Counter(B).values(), dtype=float)

    # Fill in N_ij
    N_ij = np.zeros((C_F,C_M))
    for i in range(C_F):
        for j in range(C_M):
            A_i = ((A == i))*1
            B_j = (B == j)*1
            N_ij[i,j] = sum(np.multiply(A_i,B_j))
    
    # Numerator
    num_sum = 0
    for i in range(C_F):
        for j in range(C_M):
            if N_ij[i,j] != 0:
                num_sum = num_sum + N_ij[i,j]*np.log((98*N_ij[i,j])/(N_i[i]*N_j[j]))
        
    num = -2* num_sum
    
    # Denominator
    den = sum(N_i*np.log(N_i/n_roi)) + sum(N_j*np.log(N_j/n_roi))
    
    
    nim = num / den
    
    return nim
```

```python
# Here we calculate the NMI between all 30 subjects
assigned_system_all = np.concatenate((assigned_systems_f, assigned_systems_m), axis=0) # join subject data
n_sub = 30

nmi_within_all = np.zeros((n_sub,n_sub,n_freq))
nmi_within_all_mean = np.zeros((n_freq))

for freq in range(n_freq):
    for sub1 in range(n_sub):
        for sub2 in range(n_sub):
            subject_comm1 = assigned_system_all[sub1,:,freq]
            subject_comm2 = assigned_system_all[sub2,:,freq]
            
            nmi_within_all[sub1,sub2,freq] = NMI_func(subject_comm1,subject_comm2)
    
    np.fill_diagonal(nmi_within_all[:,:,freq], 0)
    nmi_within_all_mean[freq] = sum(sum(nmi_within_all[:,:,freq])) / (n_sub*(n_sub-1))#.shape
    
print(nmi_within_all_mean)
```

### Run script for actual data

```python
n_sub = 15 # 15 male and 15 females subjects
nmi_within_females = np.zeros((n_sub,n_sub,n_freq))
nmi_within_males = np.zeros((n_sub,n_sub,n_freq))

nmi_within_females_mean = np.zeros((n_freq))
nmi_within_males_mean = np.zeros((n_freq))
nmi_within_all_mean = np.zeros((n_freq))

for freq in range(n_freq):
    for sub1 in range(n_sub):
        for sub2 in range(n_sub):
            
            F_comm1 = assigned_systems_f[sub1,:,freq]
            F_comm2 = assigned_systems_f[sub2,:,freq]
        
            M_comm1 = assigned_systems_m[sub1,:,freq]
            M_comm2 = assigned_systems_m[sub2,:,freq]

            nmi_within_females[sub1,sub2,freq] = NMI_func(F_comm1,F_comm2)
            nmi_within_males[sub1,sub2,freq] = NMI_func(M_comm1,M_comm2)
            
    # Then get the mean of the NMI 'between group' (do you exclude the 1's where they use the same graphs)
    # We exclude the 1s across the diagonal by turning them to 1s and manually calculating the mean
    np.fill_diagonal(nmi_within_females[:,:,freq], 0)
    np.fill_diagonal(nmi_within_males[:,:,freq], 0)
    
    nmi_within_females_mean[freq] = sum(sum(nmi_within_females[:,:,freq])) / (n_sub*(n_sub-1))
    nmi_within_males_mean[freq] = sum(sum(nmi_within_males[:,:,freq])) / (n_sub*(n_sub-1))
    
    # We now determine the mean nmi for all within group (i.e. within the female and male group separately)
    nmi_within_all_mean[freq] = (nmi_within_females_mean[freq] + nmi_within_males_mean[freq]) / 2
            
np.save('../../../Data/06-Sex_analyses/03-Static_modular_analyses/nmi_within_females_mean.npy', nmi_within_females_mean)
np.save('../../../Data/06-Sex_analyses/03-Static_modular_analyses/nmi_within_males_mean.npy', nmi_within_males_mean)     
```

### Run 10000 permutations

```python
n_perm = 1000
p_value_comm_f = np.zeros((n_freq))
p_value_comm_m = np.zeros((n_freq))
p_value_comm_all = np.zeros((n_freq))

nmi_within_females_perm = np.zeros((n_perm,n_sub,n_sub,n_freq))
nmi_within_males_perm = np.zeros((n_perm,n_sub,n_sub,n_freq))

nmi_within_females_perm_mean = np.zeros((n_perm,n_freq))
nmi_within_males_perm_mean = np.zeros((n_perm,n_freq))
nmi_within_all_perm_mean = np.zeros((n_perm,n_freq))

# Now we want to randomise the male and female data
# We do this by randomizing indexes, and then randomising along that dimension in a combined assigned_systems_ f + m
idx = np.array(range(30))
idx_shuffle = idx.copy()
combined_data = np.vstack((assigned_systems_f,assigned_systems_m))

for freq in range(n_freq):
    print(freq)
    combined_data_copy = combined_data[:,:,freq].copy()
    random.seed(123) # set seed for reproducibility

    # Perform randomization 10000 times
    for perm in range(n_perm):
        random.shuffle(idx_shuffle)
        combined_data_copy = combined_data_copy[idx_shuffle,:]
        shuffled_female = combined_data_copy[0:15,:]
        shuffled_male = combined_data_copy[15:30,:]
        
        
        for sub1 in range(n_sub):
            for sub2 in range(n_sub):
                F_comm1 = shuffled_female[sub1,:]
                F_comm2 = shuffled_female[sub2,:]
        
                M_comm1 = shuffled_male[sub1,:]
                M_comm2 = shuffled_male[sub2,:]

                nmi_within_females_perm[perm,sub1,sub2,freq] = NMI_func(F_comm1,F_comm2)
                nmi_within_males_perm[perm,sub1,sub2,freq] = NMI_func(M_comm1,M_comm2)
                
                np.fill_diagonal(nmi_within_females_perm[perm,:,:,freq], 0)
                np.fill_diagonal(nmi_within_males_perm[perm,:,:,freq], 0)
                
        
        nmi_within_females_perm_mean[perm,freq] = sum(sum(nmi_within_females_perm[perm,:,:,freq])) / (n_sub*(n_sub-1))
        nmi_within_males_perm_mean[perm,freq] = sum(sum(nmi_within_males_perm[perm,:,:,freq])) / (n_sub*(n_sub-1))
        
        nmi_within_all_perm_mean[perm,freq] = (nmi_within_females_perm_mean[perm,freq] + nmi_within_males_perm_mean[perm,freq])/2
        
        
    # Calc number of permutations that are greater than the value calculated earlier and divide this by number 
    # of permutations to get the p-values   
    p_value_comm_f[freq] = (sum(i > nmi_within_females_mean[freq] for i in nmi_within_females_perm_mean[:,freq]))/n_perm
    p_value_comm_m[freq] = (sum(i > nmi_within_males_mean[freq] for i in nmi_within_males_perm_mean[:,freq]))/n_perm
    p_value_comm_all[freq] = (sum(i > nmi_within_all_mean[freq] for i in nmi_within_all_perm_mean[:,freq]))/n_perm
```
