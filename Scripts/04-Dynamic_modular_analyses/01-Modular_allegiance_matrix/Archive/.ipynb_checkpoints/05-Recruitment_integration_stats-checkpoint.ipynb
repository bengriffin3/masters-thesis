### Step 2: MLM modeling for all networks

```python
print(stats)
#summary(stats)
#anova(w,x,y,z) computes the AIC, BIX, loglik for each of w,x,y,z. It then tests the models by: w vs x, x vs y, y vs z
#non-significant p-value --> use model with smaller df; significant p-value --> use model with smaller AIC –
# the df value tell you which model is more complex - it is number of parameters, which is the degrees of freedom!!
# the L-ratio is chi?

# This shows us the df (degrees of freedom - no. of parameters), sigma (residual standard error), logLik, AIC, and BIC for each model
glance(baseline)
glance(Freqband)
glance(sex)
glance(Freqband_sex)

#AIC stands for (Akaike’s Information Criteria). The basic idea of AIC is to penalize the inclusion of additional 
#variables to a model. It adds a penalty that increases the error when including additional terms. 
# The lower the AIC, the better the model.

# AIC = −2 log-likelihood + 2n_par, where n_par = number of paramters
# Therefore number of parameters:
# baseline = 8/2 = 4
# Freqband = 18/2 = 9
# sex = 20/2 = 10
# Freqband_sex  = 30/2 = 15
# This is the same as df!

#BIC (or Bayesian information criteria) is a variant of AIC with a stronger penalty for including additional 
# variables to the model.

# BIC = -2 log-likelihood + k * enp, where k = log(n)
 
```

```python
# create blank data frames
mlm_stats <- data.frame() 
mlm_params <- data.frame()

for (i in 1:length(unique_networks)){
    for (j in i:length(unique_networks)){
        network1 <- as.character(unique_networks[i]) # assign network 1 to be a system ("AU" or "VIS" etc.)
        network2 <- as.character(unique_networks[j])# assign network 2 to be a system ("AU" or "VIS" etc.)
    
        net <- networks_allegiance %>% filter(Network == network1) # filter the whole table for only system 1
        variable <- net[[network2]] # filter the remaining system 1 by the corresponding system 2
        
        # at this point, variable is only the inputs corresponding to system 1 for the rows and system 2 for the columns


        baseline <- lme(variable ~ 1, random = ~ 1 |Subject/Freqband, data = net, method = 'ML',  control=lmeControl(returnObject=TRUE))#control = list(opt = "optim"))
        # This generic function fits a linear mixed-effects model in the formulation
        # A mixed-effects model is a statistical model containing both fixed effects and random effects
        # variable ~ 1 decribes the fixed-effects part of the model
        # random ~1|Subject/Freqband is the random part of the model
        # The notation / means that Freqband is nested in Subject. (I think this means they are both random)
        # The 1 in random=~1 indicates that an intercept is to be fitted for each level of the random variable.
        # 'variable ~ 1' is a formula (e.g. like y = x + c would be written as y ~ x + c), where variable is the dependent 
        # variable, and 1 is the independent variable
        # method = 'ML' means the log-likelihood is maximized (non restricted)
        
        Freqband <- update(baseline, .~. + Freqband) # this is a model for frequency bands
        # in order to modify formulae without converting them to character you can use the update() function
        # baseline is the current model, now we change the formula by adding Freqband
        sex <- update(Freqband, .~. + Sex) # this is a model for sex
        Freqband_sex <- update(sex, .~. + Freqband:Sex)# this is a model for frequency bands / sex
        
        # The     .~. + x means that the left side of the model is kept the same and the right side is updated with x
        
        # : means interaction i.e. Freqband interacts with Sex, which means that you only want
        # the variables' interaction, and not the variables themselves
        
        # print which networks we are comparing
        print(paste(network1, network2))
        
        # Up to this point, we have updated the models so we have 4 models, now we want to statistically analyse them
        # We do this by analysing baseline, and then comparing Freqband to baseline, sex to Freqband, and Freqband_sex to sex
        # Note: When given a sequence of objects, anova tests the models against one another in the order specified.
        stats <- anova(baseline, Freqband, sex, Freqband_sex)
        # anova tests compares the models by using likelihood ratios (L-ratios)
        # the test is the results are significant (p-values)
        print(stats)
        # Note that the L-ratio is 2x the difference in LOG likelihoods between the models compared
        # The likelihood-ratio test requires that the models be nested – i.e. the more complex model can be transformed into the simpler model by imposing constraints on the former's parameters.
        # for more info: https://stats.stackexchange.com/questions/169524/what-are-the-desirable-statistical-properties-of-the-likelihood-ratio-test/169600#169600
        
        ### Up to this point, we have performed the statistical analysis, now we want to put it into a dataframe.
        ### We begin by summarising the models, so that we can pick out the desired results
        params_freqband <- summary(Freqband)
        params_interaction <- summary(Freqband_sex)
        
        
        # We now put the results into a (row) vector, which will be added to the data frame
        # i.e. for each i and j (pair of systems), note down the L.Ratios (chi-squared??) and p-values found
        # Note that I am currently picking the values frmo Karolina's code so need to understand which values to pick
        # if I am to use this
        vector_stats <- c(i, j, 
                          stats['L.Ratio'][2,], stats['L.Ratio'][3,], stats['L.Ratio'][4,], 
                          stats['p-value'][2,], stats['p-value'][3,], stats['p-value'][4,])
        
        vector_params <- c(i, j,
                           params_freqband$tTable[2,1], params_freqband$tTable[3,1], params_freqband$tTable[4,1], # betas
                           params_freqband$tTable[2,4], params_freqband$tTable[3,4], params_freqband$tTable[4,4], # ttests
                           params_freqband$tTable[2,5], params_freqband$tTable[3,5], params_freqband$tTable[4,5], # pvals
                           params_interaction$tTable[6,1], params_interaction$tTable[7,1], params_interaction$tTable[8,1], # betas
                           params_interaction$tTable[6,4], params_interaction$tTable[7,4], params_interaction$tTable[8,4], # ttests
                           params_interaction$tTable[6,5], params_interaction$tTable[7,5], params_interaction$tTable[8,5]) # pvals
        # Q: What are the beta values? 
        
        
        # each time through the loop we add the new (row) to eac of the data frames
        mlm_stats <- rbind(mlm_stats, vector_stats)       
        mlm_params <- rbind(mlm_params, vector_params)
     }
}
# Create column names for the data frames
colnames(mlm_stats) <- c('sys1', 'sys2', 
                         'chi_Freqband', 'chi_sex', 'chi_interaction', 
                         'pval_Freqband', 'pval_sex', 'pval_interaction')

colnames(mlm_params) <- c('sys1', 'sys2', 
                          'beta12_freqband', 'beta13_freqband', 'beta14_freqband', 
                          'ttest12_freqband', 'ttest13_freqband', 'ttest14_freqband',
                          'pval12_freqband', 'pval13_freqband', 'pval14_freqband',
                          'beta12_interaction', 'beta13_interaction', 'beta14_interaction', 
                          'ttest12_interaction', 'ttest13_interaction', 'ttest14_interaction',
                          'pval12_interaction', 'pval13_interaction', 'pval14_interaction')


```

```python
library(emmeans) # emmeans: Estimated Marginal Means, aka Least-Squares Means

network1 <- 'DM' # let's take a deeper look at DM
network2 <- 'FP' # and FP
net = networks_allegiance %>% filter(Network == network1) # Filter the rows so we only have DM
variable <- net[[network2]] # Filter the columns so we only have FP

# We do exactly as we did before but only for DM / FP
baseline <- lme(variable ~ 1, random = ~ 1 |Subject/Freqband, data = net, method = 'ML',  control=lmeControl(returnObject=TRUE))
Freqband <- update(baseline, .~. + Freqband)
sex <- update(Freqband, .~. + Sex)
Freqband_sex <- update(sex, .~. + Freqband:Sex)

anova(baseline, Freqband, sex, Freqband_sex)
summary(Freqband_sex)



contrast(emmeans(Freqband_sex, ~ Freqband*Sex),
         interaction = c("consec"), adjust = "bonferroni")
# emmeans(Freqband_sex, ~ Freqband*Sex) - calculates the 'Estimated Marginal Means' for each freq band in relation to both sexes
# adjust = bonferroni - correction for multiple comparisons

emmeans(Freqband_sex, pairwise ~ Freqband|Sex, adjust = "bonferroni")
```

### Sex comparison within frequency band: Delta

```python
delta_pval <- data.frame()
delta_freqband <- networks_allegiance %>% filter(Freqband=='Delta') 

for (i in 1:length(unique_networks)){
    for (j in i:length(unique_networks)){
        
        network1 <- as.character(unique_networks[i])
        network2 <- as.character(unique_networks[j])
    
        net <- delta_freqband %>% filter(Network == network1)

        stat <- t.test(net[[network2]] ~ net$Sex, paired = FALSE)
        x <- c(i, j, stat$p.value, stat$statistic)
        #print(x)
        
        delta_pval <- rbind(delta_pval, x)
        colnames(delta_pval) <- c('i', 'j', 'pval', 't')
        
        }
}

delta_pval$net1 <- unique_networks[delta_pval$i] 
delta_pval$net2 <- unique_networks[delta_pval$j]
delta_pval$pval_fdr <- p.adjust(delta_pval$pval, method = 'fdr')

# Only show us the significnat ones
delta_pval %>% filter (pval < 0.05)
```
